```python
import dotenv
import pandas as pd
from eodhd import APIClient
import os
dotenv.load_dotenv(dotenv_path="../.env")
```

```python
api_key = os.environ.get("EODHD_API_KEY")
api = APIClient(api_key)
```

```python
resp = api.get_eod_historical_stock_market_data(symbol = 'AAPL', period='d', from_date = '2023-01-01', to_date = '2023-01-15', order='a')
print(resp)
```

```python
pd.DataFrame(resp)
```

```python
import requests

url = f'https://eodhd.com/api/exchanges-list/?api_token={api_key}&fmt=json'
data = requests.get(url).json()

print(data)
```

```python
data[0]
```

```python
for item in data:
    print(item['Name'])
```
